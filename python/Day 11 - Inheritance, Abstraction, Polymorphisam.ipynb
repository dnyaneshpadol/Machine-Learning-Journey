# Inheritance:
   Inheritance allows you to reuse code by creating a new class based on an existing class.where one class (child/subclass) gets the properties and methods of another class (parent/superclass).
## Types of Inheritance in Python:
 1. Single Inheritance: One child, one parent.
 2. Multiple Inheritance: Child inherits from multiple parents.
 3. Multilevel Inheritance:Chain of inheritance.
 4. Hierarchical Inheritance: Multiple children from one parent.
 5. Hybrid Inheritance: Combination of multiple types.  

```python
# single inheritance example
class Parent:
    def show(self):
        print("this is parent class")
```

```python
class Child(Parent):
    def display(self):
        print("this is child class")
```

```python
obj = Child()
obj.show()
obj.display()
```

```python
# multiple inheritance
class Father:
    def put(self,f):
        self.father=f
```

```python
class Mother:
    def set(self,m):
        self.mother=m
```

```python
class Child(Father, Mother):
    def let(self,c):
        self.child=c
    def show(self):
        print("Father:",self.father)
        print("Mother:",self.mother)
        print("Child:",self.child)
        
```

```python
obj = Child()
obj.put('ramesh')
obj.set('kavita')
obj.let('arun')
obj.show()
```

```python
# multiplevel Inheritance
class A:
    a = 10
```

```python
class B(A):
    b = 20
```

```python
class add(B):
    c=A.a+B.b
    def show(self):
        print('addition:',self.c)
```

```python
obj = add()
obj.show()
```

```python
# Hierarchical Inheritance
class Animal:
    def eat(self):
        print("Animals eat food")
```

```python
class Dog(Animal):
    def bark(self):
        print("Dog barks")
```

```python
class Cat(Animal):
    def meow(self):
        print("Cat meows")
```

```python
d = Dog()
c = Cat()

d.eat()
d.bark()
c.eat()
c.meow()
```

```python
# Hybrid Inheritance
class A:
    def method_A(self):
        print("Class A")
```

```python
class B(A):
    def method_B(self):
        print("Class B")
```

```python
class C(A):
    def method_C(self):
        print("Class C")
```

```python
class D(B, C):
    def method_D(self):
        print("Class D")
```

```python
obj = D()
obj.method_A()
obj.method_B()
obj.method_C()
obj.method_D()
```

# overriding:
  Method overriding means a child class provides its own version of a method that already exists in the parent class.
* Child method overrides parent method
* super(): super() is used to call parent class method from child class.

```python
class Parent:
    def show(self):
        print("parent class")
```

```python
class Child(Parent):
    def show(self):
        print("child class")
```

```python
obj = Child()
obj.show()
```

```python
# useing super()
class Parent:
    def show(self):
        print("parent class")

class Child(Parent):
    def show(self):
         super().show()
         print("child class")
        
obj = Child()
obj.show()                
```

# Abstraction:
  Abstraction is an Object-Oriented Programming (OOP) concept that focuses on hiding internal implementation details and exposing only the necessary functionality to the user.
## Abstraction is implemented using:
* Abstract Classes:
  1. A class that cannot be instantiated (object cannot be created).
  2. It may contain: Abstract methods (without body), Normal methods (with body). 
* Abstract Methods:
  1. A method that is declared but not implemented.
  2. Defined using @abstractmethod decorator.
  3. Must be implemented in child class.
     
* These are provided by the built-in abc (Abstract Base Class) module.

```python
from abc import ABC, abstractmethod
```

```python
class Animal(ABC):       # abstract calss
    @abstractmethod
    def sound(self):      # abstract method
        pass
```

```python
class Dog (Animal):
    def sound(self):
        print("Dog barks")
```

```python
class Cat (Animal):
    def sound(self):
        print("Cat meows")
```

```python
d = Dog()
d.sound()
```

```python
c = Cat ()
c.sound()
```

# Polymorphism:
     it refers to the ability of a single function, method, or operator to behave differently depending on the context.
## Function Overloading () :
        Python fully supports polymorphism, but does not support traditional function overloading directly. However, overloading can be achieved using default arguments and variable-length arguments.”

```python
# function overloading using default arguments
class add:
    def sum(self,a,b,c=0):
        return a+b+c
```

```python
obj = add()
```

```python
print(obj.sum(10,20))
```

```python
print(obj.sum(10,20,30))
```

```python
print(obj.sum(7.4,55.3,22.22))
```

```python
# Using *args
class add():
    def sum(self,*args):
        return sum(args)
```

```python
obj = add()
```

```python
print(obj.sum(10,20))
```

```python
print(obj.sum(10,20,30,40,50,60,70,80))
```
