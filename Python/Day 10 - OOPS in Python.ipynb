# 4. Instance Variable:Variables that belong to an object.

```python
# examples
class car:
    def __init__(self,window,door,engine): # Constructor
        self.window=window  # variable
        self.door=door      # variable
        self.engine=engine  # variable
```

```python
car1=car(4,5,'petrol')
car2=car(3,1,'dizel')
```

```python
print(car1.window)  ## Accessing properties
```

```python
print(car1.engine)   ## Accessing properties
```

```python
print(car2.door)    ## Accessing properties
```

```python
print(car2.engine)  ## Accessing properties
```

# 5. Method: A function defined inside a class.

```python
# example 2
class test:
    def __init__(self,n,r,a):
        self.Name=n
        self.Roll=r
        self.Age=a
    def show(self):                           #method
        print("Name:{}".format(self.Name))
        print("Roll no:{}".format(self.Roll))
        print("Age:{}".format(self.Age))
```

```python
obj=test('Dnyanesh',101,21)
```

```python
obj.show()   # Calling method
```

# Access Specifiers :
   Access specifiers control the visibility and accessibility of variables and methods in a class.
 In simple termst hey decide who can access what.
1. Public Access Specifier:
Members are accessible from anywhere (inside or outside the class).
2. Protected Access Specifier (_):
Members are meant to be accessed within the class and its subclasses.
3. Private Access Specifier (__): 
Members are not accessible directly outside the class.   


```python
# all the class variable are public
class car():
    def __init__(self):
        self.name="avi"
        self.id="101"
```

```python
obj=car()
```

```python
print(obj.name)
print(obj.id)
```

```python
# all the class variable are protected
class car():
    def __init__(self):
        self._name="avi"
        self._id="101"
```

```python
obj=car()
```

```python
print(obj._name)
print(obj._id)
```

```python
# all the class variable are private
class car():
    def __init__(self):
        self.__name="avi"
        self.__id="101"
```

```python
obj=car()
```

```python
print(obj.__name) # private variable not accesible outside class
print(obj.__id)
```
