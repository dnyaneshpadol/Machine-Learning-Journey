# Pandas
 Pandas is a powerful Python library for data manipulation and analysis. It provides easy-to-use data structures and data analysis tools for working with structured data.
## Installation:
pip install pandas
## Importing Pandas:
import pandas as pd


## Real-Life Problems Pandas Solves:
### Problem 1: Working with Large Data Files
Without Pandas:
- Opening a CSV file with 10,000 rows in Excel = slow and crashes
- Hard to process thousands of records

With Pandas:
- Can easily handle millions of rows
- Fast and efficient
- Process data in seconds
### Problem 2: Data Cleaning
Without Pandas:
- You have student data with missing marks
- Manually finding and filling empty cells in Excel = tedious
- Removing duplicate entries one by one = time-consuming

With Pandas:
- Find all missing values in one line
- Fill or remove them automatically
- Remove duplicates instantly
### Problem 3: Data Analysis
Without Pandas:
- Calculate average marks of 1000 students manually
- Find who scored highest in each subject
- Group students by city and calculate stats

With Pandas:
- Calculate average, max, min in one line
- Group and analyze data easily
- Get insights quickly
### Problem 4: Reading Different File Types
Without Pandas:
- Different code to read CSV, Excel, JSON
- Complex parsing required
- Error-prone

With Pandas:
- Read CSV: pd.read_csv('file.csv')
- Read Excel: pd.read_excel('file.xlsx')
- Read JSON: pd.read_json('file.json')
- Simple and easy

# StringIO
  StringIO converts a string → file-like object
  or we can say makes string behave like file.

```python
from io import StringIO

data=('a,b,c,d\n'
           '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11,12')
```

```python
print(data)
```

```python
df=pd.read_csv(StringIO(data))
```

```python
df.head()
```

```python
# dtype parameter of read_csv 
#It controls the data type of columns while reading the CSV.
df=pd.read_csv(StringIO(data),dtype=float)
```

```python
df
```

```python
df=pd.read_csv(StringIO(data),dtype=int)
```

```python
df
```

```python
df=pd.read_csv(StringIO(data),dtype={'a':int,'b':float,'c':object})
```

```python
df
```

```python
df.dtypes
```

```python
df
```

```python
# index_col parameter of read_csv()
# making any column to index of dataset
# syntax : index_col="column_index"
#for ex consider above dataset

pd.read_csv(StringIO(data),index_col=2) # here column 2 means c will become index 
```
