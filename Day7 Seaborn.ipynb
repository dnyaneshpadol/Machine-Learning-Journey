# Seaborn
  Seaborn is a Python library used for data visualization—it helps you create beautiful and easy-to-understand graphs.

- Matplotlib is basic drawing,
- Seaborn is professional design with colors and style

## import seaborn as sns
  
## Common Graphs in Seaborn
1. Line Plot
- Shows trends over time
- Example: Sales over months

2. Bar Plot
- Compares categories
- Example: Students marks in subjects

3. Histogram
- Shows data distribution
- Example: Age distribution

4. Scatter Plot
- Shows relationship between two variables
- Example: Height vs Weight

5. Heatmap
- Shows data using colors
- Example: Correlation matrix
  

```python
import seaborn as sns
```

# Numerical Plots
1. Histogram (displot)
2. jointplot
3. pairplot

# displot
 - displot → "distribution"
 - Shows how data is distributed (spread)
 - single variable (feature)

```python
df = sns.load_dataset("tips") # tips is inbuilt dataset in seaborn
```

```python
df.head()
```

```python
sns.displot(df,x='total_bill')
```

```python
import pandas as pd
data=pd.DataFrame([102,20,130,222,123,412,53,322])
sns.displot(data)
```

# correlation
 - A correlation is a graph that shows how variables are related to each other
 - if value is 1 then both increase together
 - if value is 0 then no relation
 - if value is -1 then one increases, other decreases 

```python
df.dtypes
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

# jointplot
-  Shows relationship between 2 variables + their distributions
-  two variable(feature)
- It gives you:
1. Scatter plot (middle)
2. Histogram on X-axis
3. Histogram on Y-axis

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='hex')#The value for `kind` must be one of ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='reg')
```

# pair plot
- Shows relationships between multiple variables
- Every variable compared with every other variable
- Many Variables (Use Case:All relationships in dataset)

```python
sns.pairplot(df)
```

```python
# hue parameter
sns.pairplot(df,hue='sex') #Pairplot shows how columns are related, and hue helps compare groups (like male vs female).
```

# categorical plots
1. bosplot
2. countplot
3. bar plot
4. violinplot

# count plot
Count how many in each category

```python
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(y='day',data=df)
```

# bar plot
  Compare values between categories

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(y='total_bill',x='smoker',data=df)
```

# box plot
  Shows spread + outliers

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categories data based on smoker
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# violinplot
- A violin plot shows:
Data distribution (shape),
Spread,
Median
- It is a mix of:
Box plot and
Histogram 

```python
sns.violinplot(x='sex', y='total_bill', data=df)
```

```python
sns.violinplot(x='day', y='total_bill', data=df,palette='rainbow')
```
