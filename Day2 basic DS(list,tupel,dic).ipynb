# Boolean

# list

A list is a built-in, mutable (changeable), ordered (maintains insertion order), indexed (accessed by position starting from 0), and dynamic (can grow or shrink) collection data type in Python. Lists can store multiple items of any data type (heterogeneous), including other lists (nested). They are defined using square brackets [] with items separated by commas. Lists are iterable (can be looped over), allow duplicate values, and support both positive indexing (0, 1, 2...) and negative indexing (-1, -2, -3... from the end). They consume contiguous memory and are implemented as dynamic arrays internally.

# Creating lists:

 ## Empty list
 list_name = []

 ## List with items
 list_name = [item1, item2, item3]

 ## Mixed data types
 list_name = [1, "hello", 3.14, True]

 ## Using list() constructor
 list_name = list(iterable)

 ## Nested lists
 list_name = [[1, 2], [3, 4], [5, 6]]

 # methods
 
1. list_name.append(item)           # Add item at end
2. list_name.insert(index, item)    # Insert at position
3. list_name.remove(item)           # Remove specific item
4. list_name.pop(index)             # Remove and return item
5. list_name.extend(iterable)       # Add multiple items
6. list_name.sort()                 # Sort in place
7. list_name.reverse()              # Reverse in place
7. len(list_name)                   # Get length
8. item in list_name                # Check membership

# Accessing elements:
   
1. list_name[index]          # Access by index (0-based)
2. list_name[start:end]      # Slicing
3. list_name[-1]             # Negative indexing (last item)

# set
 A set is a collection of unique items with no duplicates and no specific order.
## Key Features:
No duplicates - automatically removes repeated values
Unordered - items have no index, can't access by position
Mutable - can add/remove items (but items inside must be immutable)
Fast lookup - checking if item exists is very quick

## syntax 
  1. my_set = {1, 2, 3, 4}
  2. empty_set = set()  # NOT {}

## Common Operations:

 1. add() - add one item
 2. remove() - remove item (error if not found)
 3. discard() - remove item (no error if not found)
 4. Union: set1 | set2 or set1.union(set2)
 5. Intersection: set1 & set2 or set1.intersection(set2)
 6. Difference: set1 - set2 or set1.difference(set2)



```python
set_var = {1,2,3,4,5,3,2}
print(set_var)
type(set_var)
```

```python
## inbuilt functions
set_var.add(6)
print(set_var)
```

```python
set_var.remove(5)
print(set_var)
```

```python
set_var.discard(6)
print(set_var)
```

```python
s1={1,2,3,4,5}
s2={7,3,4,2,1}
s1.difference(s2)
```

```python
s2.difference(s1)
```

```python
s1.union(s2)
```

```python
s1 & s2 # s1.intersection(s2)
```

# Dictionary
 A dictionary is a collection of key-value pairs where each key is unique and maps to a specific value.
 
## Key Features:
1. Key-Value pairs - stores data as key: value
2. Unordered (Python 3.7+ maintains insertion order, but not guaranteed)
3. Mutable - can add, remove, or change items
4. Keys must be unique - duplicate keys are not allowed
5. Keys must be immutable - strings, numbers, tuples (not lists)
6. Fast lookup - finding values by key is very quick

## Syntax:
1. my_dict = {"name": "John", "age": 25, "city": "Mumbai"}
2. empty_dict = {}  # or
3. dict() # inbuild function 

## Accessing Values:
1. my_dict["name"]        # Returns "John"
2. my_dict.get("age")     # Returns 25 (safer, no error if key missing)

## Common Operations:
1. dict[key] = value - add or update key-value pair
2. del dict[key] - delete a key-value pair
3. dict.keys() - get all keys
4. dict.values() - get all values
5. dict.items() - get all key-value pairs
6. dict.pop(key) - remove and return value
7. dict.update(other_dict) - merge dictionaries
8. key in dict - check if key exists

# Tuples
A tuple is an ordered, immutable (unchangeable) collection of items in Python.
## Key Features:
1. Ordered - maintains insertion order, items have fixed positions
2. Immutable - cannot add, remove, or modify items after creation
3. Indexed - access items by position (0-based indexing)
4. Allows duplicates - can have repeated values
5. Heterogeneous - can store different data types together
6. Faster than lists - due to immutability
7. Hashable - can be used as dictionary keys (unlike lists)

## Syntax:
1. my_tuple = (1, 2, 3, 4)
2. empty_tuple = ()  # or * tuple()
3. single_item = (5,)  # Note: comma is required for single item
4. mixed_tuple = (1, "hello", 3.14, True)
5. without_parentheses = 1, 2, 3  # Also creates a tuple

## Accessing Elements:
1. my_tuple[index] - access by index (0-based)
2. my_tuple[start:end] - slicing
3. my_tuple[-1] - negative indexing (last item)

## Common Operations:
1. len(tuple) - get length
2. tuple.count(item) - count occurrences of item
3. tuple.index(item) - find index of first occurrence
4. item in tuple - check if item exists
5. tuple1 + tuple2 - concatenate tuples (creates new tuple)
6. tuple * n - repeat tuple n times

## When to use:
1. Store data that should not change (constants, configurations)
2. Return multiple values from a function
3. Use as dictionary keys (unlike lists)
4. Protect data from accidental modification
5. Faster iteration when you don't need to modify data
6. Store related but different types of data together (like coordinates, RGB colors)
