# Function:
  A function in Python is a reusable block of code that performs a specific task. Instead of writing the same code again and again, you can “call” a function whenever you need it.
## syntax:
  def function_name(parameters):
    # block of code
    return value
## Types of Functions:
1. Built-in Functions:- Already available in Python{print()len()type()}
2. User-defined Functions:- Created by you using 'def'
3. Lambda Functions (Anonymous): A lambda function is a small, one-line function without a                                  name. syntax: lambda arguments: expression
4. Map Function:- The map() function is used to apply a function to every item in a list                     (or iterable). syntax: map(function, iterable)
## Parameters vs Arguments:
1. Parameter → variable in function definition
2. Argument → actual value passed

# Positional Parameters
   Values are assigned based on their position.These are parameters where order matters.

```python
def test(name,age):
    print("my name is {n} and age is {a}".format(a=age,n=name))
```

```python
test('dnyanesh',18)
```

```python
test(22,'sam') # Wrong output because order changed!
```

# Keyword Parameters
  Here you pass values using parameter names, so order doesn’t matter.

```python
test(age=22,name='sam')
```

```python
test(name='dnyanesh',age=11)
```

# *args (Variable Positional Arguments)
  - *args allows you to pass multiple positional arguments (any number of values).
  - It stores them in a tuple

# **kwargs (Variable Keyword Arguments)
  - **kwargs allows you to pass multiple keyword arguments (key=value pairs)
  - It stores them in a dictionary

```python
def test(**k):
    print(k)
```

```python
test(name='dnyanesh',age=20,address='pune')
```

```python
def test2(**kwargs):
    for key,value in kwargs.items():
        print(key,":", value)       
```

# Map function
 The map() function in Python applies a given function to every item in an iterable (like a list, tuple, etc.) and returns a map object (which you can convert to a list, tuple, etc.).
## syntax:
           map(function, iterable)
1. Instead of writing a loop like this:
 -      numbers = [1, 2, 3, 4, 5]
 -      doubled = []
 -      for num in numbers:
 -         doubled.append(num * 2)
2. You can use map() to do it in one line:
  -     numbers = [1, 2, 3, 4, 5]
   -    doubled = list(map(lambda x: x * 2, numbers))

```python
def evenodd(num):
    if num%2==0:
        return "the number {} is even".format(num)
    else:
        return "the number {} is odd".format(num)       
```

```python
lst=[1,2,3,4,5,6,7,8,9,0]
```

```python
list(map(evenodd,lst)) #returns a map object converted to list
```

```python
tuple(map(evenodd,lst)) # return map object converted to tuple
```

# lambda Function
   Lambda functions are small, anonymous (unnamed) functions in Python that you can define in a single line. They're perfect for simple operations where you don't want to write a full function definition.
 ## syntax:
    lambda arguments: expression
 1. Instead of writing:
  -            def add(x, y):
  -                return x + y
  -            result = add(5, 3)  # 8    
 2. You can write:
     -         add = lambda x, y: x + y
     -         result = add(5, 3)  # 8
 # rules
1. Anonymous: No name required (though you can assign to a variable)
2. Single expression: Only one expression, which is automatically returned
3. No return keyword: The expression result is returned automatically
4. One line: Entire function in one line

```python
def add(x,y):
    return x+y
print(add(5,4))    
```

```python
add = lambda x,y:x+y
print(add(5,4))
```
