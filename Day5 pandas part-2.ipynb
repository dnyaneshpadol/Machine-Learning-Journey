# Read Json to csv with pandas
 JSON is just a way to represent data in text form using:
- Key–value pairs
- Easy to read & write
- Used in almost every API and application

```python
import pandas as pd
```

```python
data='[{"name":"Dnyanesh","age":25},{"name":"yash","age":27}]'
```

```python
pd.read_json(data)
```

```python
from io import StringIO
```

# Reading HTML content
 In Pandas, HTML tables can be directly extracted from web pages using the function:
 - pd.read_html()
 - Reads tables from HTML pages (like websites or HTML files).
 - Automatically finds <table> tags and converts them into DataFrames.

```python
url = "https://www.w3schools.com/html/html_tables.asp"
```

```python
dfs = pd.read_html(url)
```

```python
type(dfs)
```

```python
dfs[0] # table one 
```

```python
dfs[1]  # table two
```

# Pickling 
  Pickling is the process of converting a Python object into a byte stream (binary format) so it can be saved to a file or transferred.
## Why Pickling is Used
- Save trained machine learning models
- Store complex data (lists, dictionaries, DataFrames)
- Transfer data between programs

- Pickling → object → byte stream
- Unpickling → byte stream → object

```python
# let convert above file to pickle
dfex.to_pickle('testpickle')
```

```python
df=pd.read_pickle('testpickle')
```
