# matplotlib
 Matplotlib is a Python library that helps you create graphs and charts.
Think of it like a digital drawing tool specifically made for visualizing data. Instead of manually drawing bar charts or line graphs on paper, you write a few lines of code and Matplotlib draws them for you.

- import matplotlib.pyplot as plt
  - matplotlib = the whole library
  - pyplot = a specific module inside matplotlib that provides MATLAB-style plotting

```python
import matplotlib.pyplot as plt

%matplotlib inline 
#execute code without plot.show()
```

```python
import numpy as np
```

```python
# simple example
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
## ploting using matplotlib
## plt.plot() :Showing trends over time or continuous data
plt.plot(x,y,c='r')
```

```python
# plt.scatter : Showing relationship between two variables

plt.scatter(x,y,c='b')

# giving lable to x and y axix
plt.xlabel('x axix')
plt.ylabel('y axix')
plt.title('sample graph')
plt.savefig('test.png') # to save graph
```

```python
plt.plot(x,y,c="r",linestyle='solid',linewidth=5,markersize=20)
plt.xlabel('X axix')
plt.ylabel('Y axix')
plt.title('sample data')
plt.show()
```

# plt.subplot() :Subplots let you create multiple graphs in one figure.
## plt.subplot(rows, columns, position)
- rows = how many rows of graphs
- columns = how many columns of graphs
- position = which spot to put this graph (counting left-to-right, top-to-bottom)

```python
plt.subplot(3,2,1)
plt.plot(x,y,"r--")
plt.subplot(3,2,2)
plt.plot(x,y,"b*-")
plt.subplot(3,2,3)
plt.plot(x,y,"bo")
plt.subplot(3,2,4)
plt.plot(x,y,"b*--") # due to 3 row and 2 column it will print only 6 graph and only two graph in a row
```

```python
# create sine curve
```

```python
x=np.arange(0,12,0.1)
y=np.sin(x)

#ploting
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('sine wave')
```

```python
# plot sine and cosine
x = np.arange(0,20,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# we will plot two graph in a column so row=2 and column=1
plt.subplot(2,1,1)
# first plot
plt.plot(x,y_sin,'r--')
plt.title('sine wave')

# second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'b--')
plt.title('cosine wave')
```

 # plt.bar() - Bar Chart (Vertical)
   Comparing categories

```python
# example
catgories = ['a','b','c','d']
values = [20,23,56,78]
plt.bar(catgories,values)
```

```python
# second example
c1=['a','b','c','d']
v1=[20,23,56,78]
c2=['A','B','C','D']
v2=[25,44,54,77]
 # plot bar
plt.bar(c1,v1,align='center')
plt.bar(c2,v2)
plt.show()
```

 # plt.barh() - Horizontal Bar Chart
   Same as bar chart, but horizontal (good for long category names)

```python
cat = ['product A','product B','product c']
val = [23,46,57]
plt.barh(cat,val)
plt.show()
```

# plt.boxplot() - Box Plot
  Showing data distribution with quartiles Ex,, comparing salary ranges
1. Minimum
- Smallest value in the dataset
- Left whisker end
2. Q1 (First Quartile)
- 25% of data lies below this
- Left side of the box
3. Median (Q2)
- Middle value (50%)
- Line inside the box
4. Q3 (Third Quartile)
- 75% of data lies below this
- Right side of the box
5. Maximum
- Largest value
- Right whisker end

```python
data = [7, 8, 9, 10, 11, 12, 13, 14, 15]
plt.boxplot(data)
plt.show()
```

#  plt.pie() - Pie Chart
   Showing parts of a whole (percentages) ex.,,survey results
   

```python
# example
size=[290,310,420,220]
labels=['c','c++','java','python']
plt.pie(size,labels=labels)
plt.show()
```

```python
# using different parameter
explode=[0.2,0.1,0,0]
col=['red','green','lightskyblue','yellowgreen']
plt.pie(size,explode=explode,labels=labels,colors=col,autopct='%1.1f%%',shadow=True)

# size: indicate size of labels it calculate based on %
# explode: indicates how much far slice to go as above slice one is 0.2 and 0.1
# labels : indicates labels
# colors : indicates color of slice   
# autopct : show percentage 
# shadow : it can be true or false
plt.show()
```
