# Exception:
An exception is an error that occurs during program execution.
# Why Exception Handling?
1. Prevents program crash 
2. Helps show user-friendly messages 
3. Allows graceful recovery from errors
## Types of Blocks:
1. try block: Contains risky code.
2. except block: Handles error.
3. else block: Runs if no exception occurs.
4. finally block: Always runs (whether error occurs or not).
## raise keyword:
The raise keyword in Python is used to manually generate (throw) an exception.

```python
# try-except block example
try:
    a=b
except:
    print("some error occure")
```

```python
try:
    a=b
except NameError as ex1:
    print(ex1)
except:
    print("some error occure")    
```

```python
# try-except-else block
try:
    a=int(input("enter first number"))
    b=int(input("enter second number"))
    c=a/b
    d=a*b
    e=a+b
except NameError:
    print("variable is not defined")
except ZeroDivisionError :
    print("number can not divide by zero")
except Exception as ex:
    print(ex)
else:
    print("div:",c)
    print("mul:",d)
    print("add:",e)   
finally:
    print("the execution is done")    
```

# custom Exception
 A custom exception is a user-defined error.
Instead of using only built-in errors like ValueError or TypeError, you can create your own exception class for specific situations in your program.

```python
# You can create it by inheriting from the built-in Exception class.
class Age(Exception):
    pass
```

```python
year=int(input("enter the birth year"))
age=2026-year
try:
    
    if age>18:
        print(" allowed, you are 18+")
    else:
        raise Age
except Age:
    print("not allowed, you are not 18+")
```

```python
# Multiple Custom Exceptions
class Error(Exception):
    pass
    
class NegativeNumberError(Error):
    pass

class ZeroError(Error):
    pass
```

```python
num = int(input("enter a number"))

try:
    if num < 0:
        raise NegativeNumberError("Negative not allowed")
    elif num == 0:
        raise ZeroError("Zero not allowed")
except NegativeNumberError as e:
    print(e)
except ZeroError as e:
    print(e)
```
