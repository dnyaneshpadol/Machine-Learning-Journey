# variables
A variable is a named container that stores data in memory. It acts as a label or reference to a value.

## Key Features:
1. No declaration needed - Python automatically determines type
2. Dynamic typing - can change type during execution
3. Case-sensitive - name and Name are different variables
4. Stores references to objects in memory
5. Can be reassigned to different values

## Naming Rules:
1. Must start with a letter (a-z, A-Z) or underscore (_)
2. Can contain letters, numbers, and underscores
3. Cannot start with a number
4. Cannot use Python keywords (if, for, while, etc.)
5. Cannot contain spaces or special characters (@, #, $, %, etc.)
6. Case-sensitive (age and Age are different)

## Naming Conventions:
1. Use lowercase with underscores for variable names (snake_case)
2. Use descriptive names (student_name instead of sn)
3. Constants use UPPERCASE (PI = 3.14)
4. Avoid single letters except in loops (i, j, k)

```python
a=10
print(a)
```

```python
b='abc'
print(b)
```

```python
c=10.11
print(c)
```

```python
d=True
print(d)
```

# type ()
The type() function returns the data type (class) of an object in Python.
## Syntax:
1. type(object)
## Key Features:
1. Returns the class/type of any object
2. Built-in function available everywhere
3. Helps in debugging and type checking
4. Works with all Python data types
## Common Data Types Returned:
1. <class 'int'> - Integer numbers
2. <class 'float'> - Decimal numbers
3. <class 'str'> - Strings (text)
4. <class 'bool'> - Boolean (True/False)
5. <class 'list'> - Lists
6. <class 'tuple'> - Tuples
7. <class 'dict'> - Dictionaries
8. <class 'set'> - Sets
9. <class 'NoneType'> - None value

```python
print(type(1))
```

```python
print(type(str))
```

```python
print(type(10.11))
```

```python
print(type(True))
```

# assignment 
Assignment is the process of storing a value in a variable using the assignment operator (=).
## Key Features:
1. Uses the = operator to assign values
2. Right side value is assigned to left side variable
3. Creates variable if it doesn't exist
4. Updates variable if it already exists
5. Right side is evaluated first, then assigned to left

## Syntax:
1. variable_name = value
2. variable = expression
3. variable = another_variable


```python
a=10
print(a)
```

```python
a="abc xyz"
print(a)
```

```python
  # multiple assignment
a = b = c = 100  
print(a,b,c)
```

```python
a, b, c = 1, 2, 3
print(a, b, c)
```

```python
#swaping
a=10
b=20
print("begore swap",a,b)
a,b=b,a
print ("after swap ",a, b)
```

## arithmatic operator
Arithmetic operators are used to perform mathematical operations on numbers in Python
## Operator Precedence (Order of Operations):
1. ** (Exponentiation) - Highest priority
2. *, /, //, % (Multiplication, Division, Floor Division, Modulus)
3. +, - (Addition, Subtraction) - Lowest priority

```python
x=40
y=20

print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(x*(x-y)+(x+y)/y)
print(x**y) # x raised to y
```

# string formating
   String formatting is the process of inserting variables or values into a string in a structured way.
## Methods of String Formatting:  
 1. f-strings (Formatted String Literals)
 2. format() method

```python
fname="dnyanesh"
lname="padol"
```

```python
print("my first name is {} and last name is {}".format(fname,lname))
```

```python
print("my first name is {f} and last name is {l}".format(f=fname,l=lname))
```

```python
print(f"my first name is {fname} and last name is {lname}")
```

```python
print("First: {0}, Last: {1}, Full: {0} {1}".format(fname, lname))
```

# len()
The len() function returns the length (number of items) of an object.
## Syntax:
len(object)
## Key Features:
1. Returns integer value
2. Works with sequences (strings, lists, tuples)
3. Works with collections (sets, dictionaries)
4. Built-in function
5. O(1) time complexity (very fast)


```python
len('dnyanesh')
```
